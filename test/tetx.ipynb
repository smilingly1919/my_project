```python
import os
import sdf_helper as sh  # 假设你已经导入了

# 基础路径列表（可按需求增减）
base_paths = [
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-0.5微米',
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-1微米',
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-1.5微米',
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-2微米',
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-2.5微米',
    '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-3微米',
]

file_index = 100
file_name = f"distfun{file_index:04d}.sdf"

# 用列表推导批量拼路径
file_paths = [os.path.join(bp, file_name) for bp in base_paths]

# 批量加载数据，存进列表
data_list = [sh.getdata(fp) for fp in file_paths]


results = [load_distfun(fp, "Electron", "x_px") for fp in file_paths]


x0,px0,N0 = results[0]["grid_x"],results[0]["grid_y"],results[0]["dist"]
x1,px1,N1 = results[1]["grid_x"],results[1]["grid_y"],results[1]["dist"]
x2,px2,N2 = results[2]["grid_x"],results[2]["grid_y"],results[2]["dist"]
x3,px3,N3 = results[3]["grid_x"],results[3]["grid_y"],results[3]["dist"]
x4,px4,N4 = results[4]["grid_x"],results[4]["grid_y"],results[4]["dist"]
x5,px5,N5 = results[5]["grid_x"],results[5]["grid_y"],results[5]["dist"]


# 沿 px（第1维）方向求和，得到 N(x) 分布
sum0 = np.sum(N0, axis=1)
sum1 = np.sum(N1, axis=1)
sum2 = np.sum(N2, axis=1)
sum3 = np.sum(N3, axis=1)
sum4 = np.sum(N4, axis=1)
sum5 = np.sum(N5, axis=1)

# 绘图
plt.figure(figsize=(8,6))
# plt.xlim(2e-5,5e-5)
plt.plot(x0, sum0, label='a=145, 0.5μm')
plt.plot(x1, sum1, label='a=145, 1μm')
plt.plot(x2, sum2, label='a=145, 1.5μm')
plt.plot(x3, sum3, label='a=145, 2μm')
plt.plot(x4, sum4, label='a=145, 2.5μm')
plt.plot(x5, sum5, label='a=145, 3μm')

plt.xlabel('x (m)')
plt.ylabel('∑ N(x, px)')
plt.title('Integrated Distribution over px')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# load_distfun(file_paths[0], "Photon", "x_py")
# print("所有键：", result.keys())
# x = result["grid_x"]
# py = result["grid_y"]
# N = result["dist"]
```
